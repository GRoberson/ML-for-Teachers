```
# prompt: create a normal distribution curve with mean and standard deviation

import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution

# Generate x values (range of the distribution)
x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)

# Calculate the probability density function (PDF) of the normal distribution
pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Create the plot
plt.plot(x, pdf)
plt.title('Normal Distribution Curve')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)

# Display the plot
plt.show()
```

```
# prompt: create a tool that can give a control to change the mean and standard deviation

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

# Function to generate and plot the normal distribution
def plot_normal_distribution(mean, std_dev):
    x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    plt.plot(x, pdf)
    plt.title('Normal Distribution Curve')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.show()

# Create interactive sliders for mean and standard deviation
interact(plot_normal_distribution,
         mean=FloatSlider(min=-5, max=5, step=0.1, value=0),
         std_dev=FloatSlider(min=0.1, max=5, step=0.1, value=1));
```

```
# prompt: add option to calculate cdf upto value x and highlight area under the curve

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from scipy.stats import norm

# Parameters for the normal distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation of the distribution

# Function to generate and plot the normal distribution with CDF highlighting
def plot_normal_distribution(mean, std_dev, x_value):
    x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    cdf = norm.cdf(x_value, loc=mean, scale=std_dev)

    plt.plot(x, pdf)
    plt.title('Normal Distribution Curve')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)

    # Highlight area under the curve up to x_value
    x_fill = np.linspace(mean - 3 * std_dev, x_value, 100)
    plt.fill_between(x_fill, 0, norm.pdf(x_fill, loc=mean, scale=std_dev), alpha=0.3)

    plt.text(x_value + 0.1, 0.1, f"CDF(x={x_value:.2f}) = {cdf:.3f}", fontsize=10)  # Display CDF value
    plt.show()

# Create interactive sliders for mean, standard deviation, and x-value
interact(plot_normal_distribution,
         mean=FloatSlider(min=-5, max=5, step=0.1, value=0),
         std_dev=FloatSlider(min=0.1, max=5, step=0.1, value=1),
         x_value=FloatSlider(min=-5, max=5, step=0.1, value=0));
```

```
# prompt: add a slider to change x value and print the associated cdf value on graph

import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from scipy.stats import norm

# Function to generate and plot the normal distribution with CDF highlighting
def plot_normal_distribution(mean, std_dev, x_value):
    x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    cdf = norm.cdf(x_value, loc=mean, scale=std_dev)

    plt.plot(x, pdf)
    plt.title('Normal Distribution Curve')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)

    # Highlight area under the curve up to x_value
    x_fill = np.linspace(mean - 3 * std_dev, x_value, 100)
    plt.fill_between(x_fill, 0, norm.pdf(x_fill, loc=mean, scale=std_dev), alpha=0.3)

    plt.text(x_value + 0.1, 0.1, f"CDF(x={x_value:.2f}) = {cdf:.3f}", fontsize=10)  # Display CDF value
    plt.show()

# Create interactive sliders for mean, standard deviation, and x-value
interact(plot_normal_distribution,
         mean=FloatSlider(min=-5, max=5, step=0.1, value=0),
         std_dev=FloatSlider(min=0.1, max=5, step=0.1, value=1),
         x_value=FloatSlider(min=-5, max=5, step=0.1, value=0));
```

```
# prompt: add one another x_value to get area between two x_values

# Function to generate and plot the normal distribution with CDF highlighting between two x-values
def plot_normal_distribution(mean, std_dev, x_value1, x_value2):
    x = np.linspace(mean - 3 * std_dev, mean + 3 * std_dev, 100)
    pdf = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    cdf1 = norm.cdf(x_value1, loc=mean, scale=std_dev)
    cdf2 = norm.cdf(x_value2, loc=mean, scale=std_dev)
    area = cdf2 - cdf1

    plt.plot(x, pdf)
    plt.title('Normal Distribution Curve')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)

    # Highlight area under the curve between x_value1 and x_value2
    x_fill = np.linspace(x_value1, x_value2, 100)
    plt.fill_between(x_fill, 0, norm.pdf(x_fill, loc=mean, scale=std_dev), alpha=0.3)

    plt.text((x_value1 + x_value2) / 2, 0.1, f"Area = {area:.3f}", fontsize=10)  # Display area
    plt.show()

# Create interactive sliders for mean, standard deviation, and x-values
interact(plot_normal_distribution,
         mean=FloatSlider(min=-5, max=5, step=0.1, value=0),
         std_dev=FloatSlider(min=0.1, max=5, step=0.1, value=1),
         x_value1=FloatSlider(min=-5, max=5, step=0.1, value=-1),
         x_value2=FloatSlider(min=-5, max=5, step=0.1, value=1));
```
