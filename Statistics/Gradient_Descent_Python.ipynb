## Gradient Descent

```python
# Create a scatter plot for small dataset
import matplotlib.pyplot as plt

# Create the dataset
x = [2, 4, 6]
y = [6, 8, 7]

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Data Points")

# Add point labels
plt.annotate('p1', (x[0], y[0]))
plt.annotate('p2', (x[1], y[1]))
plt.annotate('p3', (x[2], y[2]))


# Display the plot
plt.show()
```

```python
# Fitting a base line having m=1 and c=0

# Create the scatter plot
plt.scatter(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of Data Points")

# Add point labels
plt.annotate('p1', (x[0], y[0]))
plt.annotate('p2', (x[1], y[1]))
plt.annotate('p3', (x[2], y[2]))

#Plot the line
x_line = [0,7] #Define x range of the line
m=1
c=0
y_line = [m*i+c for i in x_line]
plt.plot(x_line,y_line)


# Display the plot
plt.show()
```

```python
# Calculation for sum of squared residual

import numpy as np

# Assuming x and y are defined as in your previous code
x = np.array(x)
y = np.array(y)

# Define the line parameters (you can change these)
m = 1
c = 0

# Calculate the predicted y-values
y_pred = m * x + c

# Calculate the squared residuals
squared_residuals = (y - y_pred)**2

# Calculate the sum of squared residuals
ssr = np.sum(squared_residuals)

print("Sum of Squared Residuals (SSR):", ssr)
```

```python
# Calculating the gradient for Intercept c and step size

# Define the line parameters (you can change these)
learning_rate = 0.01

# Calculate the predicted y-values
y_pred = m * x + c

# Calculate the gradient of the cost function with respect to c
gradient_c = -2 * np.sum(y - y_pred)

# Update c using gradient descent
new_c = c - learning_rate * gradient_c

print("Old value of c:", c)
print("Gradient with respect to c:", gradient_c)
print("New value of c:", new_c)

# Calculate the predicted y-values with the new c
y_pred_new = m * x + new_c

# Calculate the new sum of squared residuals
new_ssr = np.sum((y - y_pred_new)**2)

print("Old Sum of Squared Residuals (SSR):", np.sum((y - y_pred)**2))
print("New Sum of Squared Residuals (SSR):", new_ssr)
```

```python
# Calculating the gradient for slope m and step size

# Calculate the gradient of the cost function with respect to m
gradient_m = -2 * np.sum(x * (y - y_pred))

# Update m using gradient descent
new_m = m - learning_rate * gradient_m

print("Old value of m:", m)
print("Gradient with respect to m:", gradient_m)
print("New value of m:", new_m)

# The difference between new_m and m is the step size for m
step_size_m = new_m - m
print("Step size for m:", step_size_m)
```

```python
# Calculating the gradient for intercept c and slope m and step size simultaniously

# Calculate the gradient of the cost function with respect to c
gradient_c = -2 * np.sum(y - y_pred)
gradient_m = -2 * np.sum(x * (y - y_pred))

# Update c using gradient descent
new_c = c - learning_rate * gradient_c
new_m = m - learning_rate * gradient_m

print("Old value of c:", c)
print("Old value of m:", m)
print("Gradient with respect to c:", gradient_c)
print("Gradient with respect to m:", gradient_m)
print("New value of c:", new_c)
print("New value of m:", new_m)

# Calculate the predicted y-values with the new c
y_pred_new = m * x + new_c

# Calculate the new sum of squared residuals
new_ssr = np.sum((y - y_pred_new)**2)

print("Old Sum of Squared Residuals (SSR):", np.sum((y - y_pred)**2))
print("New Sum of Squared Residuals (SSR):", new_ssr)
```

```python
# Creating a gradeint fuction to optimze sum of squared residual and plotting it on a graph

# Define the line parameters (you can change these)
epochs = 10  # Number of iterations

# Lists to store m, c, and SSR values for plotting
m_values = []
c_values = []
ssr_values = []

for epoch in range(epochs):
    # Calculate the predicted y-values
    y_pred = m * x + c

    # Calculate the gradients
    gradient_c = -2 * np.sum(y - y_pred)
    gradient_m = -2 * np.sum(x * (y - y_pred))

    # Update m and c using gradient descent
    new_c = c - learning_rate * gradient_c
    new_m = m - learning_rate * gradient_m

    m = new_m
    c = new_c

    # Calculate the SSR
    ssr = np.sum((y - y_pred)**2)

    #Store values
    m_values.append(m)
    c_values.append(c)
    ssr_values.append(ssr)


# Plotting
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(range(epochs), m_values, marker='o')
plt.xlabel("Epoch")
plt.ylabel("m")
plt.title("m over Epochs")

plt.subplot(1, 3, 2)
plt.plot(range(epochs), c_values, marker='o', color='orange')
plt.xlabel("Epoch")
plt.ylabel("c")
plt.title("c over Epochs")

plt.subplot(1, 3, 3)
plt.plot(range(epochs), ssr_values, marker='o', color='green')
plt.xlabel("Epoch")
plt.ylabel("SSR")
plt.title("SSR over Epochs")

plt.tight_layout()
plt.show()
```

```python
#Plotting the final values

# Lists to store m, c, and SSR values for plotting
m_values = []
c_values = []
ssr_values = []

for epoch in range(epochs):
    # Calculate the predicted y-values
    y_pred = m * x + c

    # Calculate the gradients
    gradient_c = -2 * np.sum(y - y_pred)
    gradient_m = -2 * np.sum(x * (y - y_pred))

    # Update m and c using gradient descent
    new_c = c - learning_rate * gradient_c
    new_m = m - learning_rate * gradient_m

    m = new_m
    c = new_c

    # Calculate the SSR
    ssr = np.sum((y - y_pred)**2)

    #Store values
    m_values.append(m)
    c_values.append(c)
    ssr_values.append(ssr)

#Plot the final graph
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label='Data Points')

#Plot the final fitted line
x_line = np.linspace(min(x), max(x), 100) #More points for smoother line
y_line = m * x_line + c
plt.plot(x_line, y_line, color='red', label=f'Fitted Line (m={m:.2f}, c={c:.2f})')

#Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Final Fitted Line with Data Points")

# Add point labels
for i in range(len(x)):
  plt.annotate(f'p{i+1}', (x[i], y[i]))

plt.text(3, 4, f"SSR={ssr:.2f}")
plt.legend()
plt.show()
```
